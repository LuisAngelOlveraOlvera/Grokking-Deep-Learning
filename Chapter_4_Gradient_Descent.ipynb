```python
# An empty network
weight = 0.1
lr = 0.01

def neural_network(input, weight):
    prediction = input * weight
    return prediction

# PREDICT: Making a prediction and evaluating error
number_of_toes = [8.5]
win_or_lose_binary = [1]  # [won]

input = number_of_toes[0]
true = win_or_lose_binary[0]

pred = neural_network(input, weight)

error = (pred - true) ** 2
print("Initial error:", error)

# COMPARE: Making a prediction with a higher weight and evaluating error
lr = 0.1
p_up = neural_network(input, weight + lr)
e_up = (p_up - true) ** 2
print("Error with higher weight:", e_up)

# COMPARE: Making a prediction with a lower weight and evaluating error
lr = 0.01
p_dn = neural_network(input, weight - lr)
e_dn = (p_dn - true) ** 2
print("Error with lower weight:", e_dn)


# COMPARE + LEARN: Comparing the errors and setting the new weight
if (error > e_dn) or (error > e_up):
    if error > e_dn:
        weight -= lr
    else:
        weight += lr

print("Initial weight:", 0.1)
print("Adjusted weight:", weight)
```

```python
# Hot and cold learning
weight = 0.5
input = 0.5
goal_prediction = 0.8

step_amount = 0.001

for iteration in range(1101): # repeat the learning many times so the error can keep getting smaller
  prediction = input * weight
  error = (prediction - goal_prediction) ** 2
  print("Iteration:", iteration, "Error:", error, "Prediction:", prediction)

  up_prediction = input * (weight + step_amount) # try up!
  up_error = (goal_prediction - up_prediction) ** 2

  down_prediction = input * (weight - step_amount) # try down!
  down_error =(goal_prediction - down_prediction) ** 2

  if(down_error < up_error):
    weight = weight - step_amount  # if down is better, go down
  else:
    weight = weight + step_amount # if up is better, go up
```

```python
# Definir la funcin cuadrtica
def function(x):
    return x ** 2

# Definir la derivada de la funcin cuadrtica
def derivative(x):
    return 2 * x

# Parmetros del algoritmo de descenso de gradiente
learning_rate = 0.1  # Tasa de aprendizaje
iterations = 100  # Nmero de iteraciones
initial_guess = 4  # Suposicin inicial

# Descenso de gradiente
current_guess = initial_guess
for i in range(iterations):
    gradient = derivative(current_guess)  # Calcular el gradiente en el punto actual
    current_guess -= learning_rate * gradient  # Actualizar la suposicin utilizando el gradiente y la tasa de aprendizaje
    print(f"Iteration {i+1}: x = {current_guess}, f(x) = {function(current_guess)}")

print("Resultado final:", current_guess)
```

```python
# Same excersie than aboce but with graph to show how the point is reachin out the point
import matplotlib.pyplot as plt

weight, goal_pred, input_val = (0.0, 0.8, 1.1)
weight_history = []
error_history = []

for iteration in range(4):
    pred = input_val * weight
    error = (pred - goal_pred) ** 2
    delta = pred - goal_pred
    weight_delta = delta * input_val
    weight = weight - weight_delta

    weight_history.append(weight)
    error_history.append(error)

# Crear valores de peso para trazar la curva de error
weights_for_curve = [i / 10 for i in range(int(min(weight_history) * 10) - 1, int(max(weight_history) * 10) + 2)]

# Calcular errores correspondientes
errors_for_curve = [(input_val * w - goal_pred) ** 2 for w in weights_for_curve]

plt.plot(weights_for_curve, errors_for_curve, label='Error Curve', color='blue')
plt.scatter(weight_history, error_history, marker='o', color='red', label='Weight Movement')
plt.xlabel('Weight')
plt.ylabel('Error')
plt.title('Movement of Weight towards Minima')

# Agregar etiquetas a cada punto
for i, txt in enumerate(weight_history):
    plt.text(weight_history[i], error_history[i], f'Iteration {i+1}')

plt.grid(True)
plt.legend()
plt.show()

```
